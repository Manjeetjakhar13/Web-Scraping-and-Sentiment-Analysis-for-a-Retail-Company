# Prepare user defined function to extract data from all the given URLS

```python
urls_list =df.URL # to get list of all urls
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',}
# headers to extract data from secure website 
```

```python
def datasetPrepare(urls_list):
    df=pd.DataFrame(columns=["url","title","content"])
    for url in urls_list:
        para=[]
        title=[]
        data = requests.get(url,headers=headers)
        soup = BeautifulSoup(data.content, 'html.parser')
        title=soup.find("h1")
        paragraphs = soup.find_all('p')
        for p in paragraphs:
            p1=p.text
            para.append(p1)
        df.loc[len(df)]=[url,title,para]
        
    return df   
```

```python
df1=pd.DataFrame()
```

```python
df1=datasetPrepare(urls_list)
```

```python
df_final=df.merge(df1,left_on="URL",right_on="url",how="inner")
```

```python
def element(a):# function to extract stop words from difrrent data frame
    s=[]
    for i in a.iloc[:,0]:
        s.append(i)
    return s
```

```python
def elementd3(a):# function to extract diffrent stopword and clean then 
    s=[]
    for i in a.iloc[:,0]:
        s.extend(i.split("|"))
    return s
```

```python
stop=[]#empty list to store stopwords
```

```python
stop.extend(element(d1))
stop.extend(element(d2))
stop.extend(elementd3(d3))
stop.extend(element(d4))
stop.extend(element(d5))
stop.extend(element(d6))
stop.extend(element(d7))
```

```python
stop.extend(['SMITH','ERNST','AFGHANI','Afghanistan','HUNDRED','ABOUT','a','UNITED'])#add remaining stop word which cannot be fetched
tokenizer = ToktokTokenizer() #initialiser tokeniser
```

```python
stop=[str(x).lower() for x in stop] #convert each stopword to lower case
```

```python
def remove_stopwords(text, is_lower_case=False): # function to  remove stopword
    tokens = tokenizer.tokenize(text)
    tokens = [token.strip() for token in tokens]
    if is_lower_case:
        filtered_tokens = [token for token in tokens if token not in stop]
    else:
        filtered_tokens = [token for token in tokens if token.lower() not in stop]
    filtered_text = ' '.join(filtered_tokens)    
    return filtered_text

```

# Remove special characters

```python
def remove_special_characters(text, remove_digits=True):
    pattern = r'[^a-zA-z0-9\s]' if not remove_digits else r'[^a-zA-z\s]'
    text = re.sub(pattern, '', text)
    return text
```

```python
def simple_stemmer(text):# function for stemming
    ps = nltk.porter.PorterStemmer()
    text = ' '.join([ps.stem(word) for word in text.split()])
    return text

```

```python
def negetive_calculator(x):
    count=0
    for i in x.split():
        if i in negetivewords:
            count=count+1
    return -(count)
```

```python
out['POLARITY SCORE']=(out["POSITIVE SCORE"]-(-out["NEGATIVE SCORE"]))/((out["POSITIVE SCORE"]+(-out["NEGATIVE SCORE"]))+0.000001)
```

* SYLLABLE PER WORD

```python
import re

def count_syllables(word):
  pattern = r'[aeiouAEIOU]+'
  syllables = re.findall(pattern, word)
  if word.endswith(("es", "ed")) and len(syllables) > 1:
    syllables = syllables[:-1]
  return len(syllables)
```

```python
def count_syllables_in_text(text):
  words = text.split()
  syllable_count = 0
  for word in words:
    syllable_count += count_syllables(word)
  return syllable_count
```

```python
out["SYLLABLE PER WORD"]=df_final.clean.apply(count_syllables_in_text)/df_final.clean.apply(lambda x:len(x.split()))
```

* COMPLEX WORD COUNT

```python
def find_complex_words(text):
  words = text.split()
  complex_words = []
  for word in words:
    if count_syllables(word) > 2:
      complex_words.append(word)
  return len(complex_words)
```

```python
out["COMPLEX WORD COUNT"]=df_final.clean.apply(find_complex_words)
```

* PERCENTAGE OF COMPLEX WORDS

```python
out["PERCENTAGE OF COMPLEX WORDS"]=(out["COMPLEX WORD COUNT"]/out["WORD COUNT"])*100
```

* FOG INDEX

```python
out["FOG INDEX"]=(out["AVG NUMBER OF WORDS PER SENTENCE"]+out["PERCENTAGE OF COMPLEX WORDS"])*0.4
```

* PERSONAL PRONOUNS

```python
import re

def count_personal_pronouns(text):
  pattern = r'\b(i|you|he|she|it|we|they|me|him|her|them)\b'
  personal_pronouns=re.findall(pattern, str(text).lower())
  num_personal_pronouns = len(personal_pronouns)
  return num_personal_pronouns
```

```python
out["PERSONAL PRONOUNS"]=df_final.content.apply(count_personal_pronouns)
```

* AVG WORD LENGTH

```python
def calculate_avg_word_length(text):
  words = text.split()
  num_words = len(words)
  total_length = 0
  for word in words:
    total_length += len(word)
  avg_length = total_length / num_words
  return avg_length
```

```python
out["AVG WORD LENGTH"]=df_final.clean.apply(calculate_avg_word_length)
```
